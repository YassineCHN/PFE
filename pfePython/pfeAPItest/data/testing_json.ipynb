```python
import json

# Chemin vers votre fichier JSONL
fichier_jsonl = "ref.jsonl"

# Liste pour stocker les objets JSON
tableau_json = []

# Lecture du fichier JSONL ligne par ligne
with open(fichier_jsonl, 'r', encoding='utf-8') as f:
    for ligne in f:
        ligne = ligne.strip()
        if ligne:  # Ignorer les lignes vides
            try:
                # Parser chaque ligne comme un objet JSON
                objet = json.loads(ligne)
                tableau_json.append(objet)
            except json.JSONDecodeError as e:
                print(f"Erreur de décodage JSON : {e}")
                print(f"Ligne problématique : {ligne}")

# Convertir la liste en JSON standard (un tableau)
json_standard = json.dumps(tableau_json, ensure_ascii=False, indent=2)

# Afficher le résultat
print(json_standard)

# Si vous souhaitez sauvegarder dans un nouveau fichier
with open("stations_standard.json", 'w', encoding='utf-8') as f:
    f.write(json_standard)

print(f"\nNombre d'objets traités : {len(tableau_json)}")
```

```python
import json

# Chemin vers le fichier mal formaté
fichier_entree = "Référentiel_stations.json"
fichier_sortie_json = "fichier_corrige.json"

# Liste pour stocker les objets
objets = []

# Lire le contenu ligne par ligne
with open(fichier_entree, 'r', encoding='utf-8') as f:
    for ligne in f:
        ligne = ligne.strip()
        if not ligne:  # Ignorer les lignes vides
            continue
            
        try:
            # Parser la ligne comme JSON
            obj = json.loads(ligne)
            objets.append(obj)
        except json.JSONDecodeError as e:
            print(f"Ligne ignorée (JSON invalide): {ligne}")
            print(f"Erreur: {e}")

# Écrire dans un fichier JSON standard
with open(fichier_sortie_json, 'w', encoding='utf-8') as f:
    json.dump(objets, f, ensure_ascii=False, indent=2)

print(f"Conversion en JSON standard terminée: {fichier_sortie_json}")
print(f"Nombre d'objets traités: {len(objets)}")
```
